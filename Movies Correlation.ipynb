# Data Cleaning

```python
df['gross'] = df['gross'].fillna(0.0).astype('int64')
df['budget'] = df['budget'].fillna(0.0).astype('int64')
```

```python
#Drop any duplicates

df.drop_duplicates()
```

```python
# Order our Data a little bit to see

df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Plot budget vs gross using seaborn

sns.regplot(x="gross", y="budget", data=df, scatter_kws={"color":"blue"}, line_kws={"color":"red"})
```

```python
correlation_matrix = df.corr(method ='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
# dictionary containing the frequency for each genre
counter = Counter(df['genre'])
genres = list(counter.keys())
freq = list(counter.values())

# positions in the Y-axis for each genre
y_pos = np.arange(len(genres))

# show plot, Genres of the most popular movies of the last 30 years
plt.barh(y_pos, freq, color="purple")
plt.yticks(y_pos, genres)
plt.xlabel('Frequency')
plt.title('Genres of the most popular movies of the last 30 years')
```

```python
# Movie revenue over the years

# list of years in the dataset (1986-2016)
years = sorted(pd.unique(df['year']))

# calculate average gross for a movie in each year
avg_gross = [np.mean(df[df['year']==year]['gross'])/1000000 for year in years]

# plot and stuff
plt.plot(years, avg_gross, color="green")
plt.ylabel('Revenue (in millions)')
plt.xlabel('Year')
plt.title('Average movie revenue over the years')
```

```python
# Looking at the top 15 companies by gross revenue

CompanyGrossSum = df.groupby('company')[["gross"]].sum()

CompanyGrossSumSorted = CompanyGrossSum.sort_values('gross', ascending = False)[:15]

CompanyGrossSumSorted = CompanyGrossSumSorted['gross'].astype('int64') 

CompanyGrossSumSorted
```

```python
df_numerized = df


for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name]= df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized
```

```python
correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Movies")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
correlation_mat = df.apply(lambda x: x.factorize()[0]).corr()

corr_pairs = correlation_mat.unstack()

print(corr_pairs)
```

```python
sorted_pairs = corr_pairs.sort_values(kind="quicksort")

print(sorted_pairs)
```

```python
# We can now take a look at the ones that have a high correlation (> 0.5)

strong_pairs = sorted_pairs[abs(sorted_pairs) > 0.5]

print(strong_pairs)
```
